```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
saving_rate = 0.64
income = int(input('Your Income'))
expenses = income * (1-saving_rate)
interest_rate = float(input('Interest Rate'))/100
investment = income - expenses
annula_return = investment * interest_rate
year = 2022
withdraw_rate = int(input('Withdrawl Rate'))/100
```

```python
Year=[]
Yearly_income = []
Yearly_Expenses=[]
Yearly_Invested_Amount=[]
Total_Invested_Amount=[]
Annula_Returns=[]
```

```python
Year.append(year)
Yearly_income.append(income)
Yearly_Expenses.append(expenses)
Yearly_Invested_Amount.append(income - expenses)
Annula_Returns.append(annula_return)
Total_Invested_Amount.append(investment)
```

```python
invested_years = 30
for i in range(0,invested_years -1):
    investment = income - expenses + annula_return + investment
    annula_return = investment * interest_rate
    Year.append(year + i +1)
    Yearly_income.append(income)
    Yearly_Expenses.append(expenses)
    Yearly_Invested_Amount.append(income - expenses)
    Annula_Returns.append(annula_return)
    Total_Invested_Amount.append(investment)
```

```python
df=pd.DataFrame()
df['Year']= Year
df['Yearly_Income']= Yearly_income
df['Yearly_Expenses']= Yearly_Expenses
df['Yearly_Invested_Amount']= Yearly_Invested_Amount
df['Total_Invested_Amount']= Total_Invested_Amount
df['Annual_Returns']= Annula_Returns
Yearly_Withdrawl_Amount = np.array(Total_Invested_Amount) * withdraw_rate
df['Yearly_Withdrawl_Amount']= Yearly_Withdrawl_Amount
```

```python
df
```

```python
plt.figure(figsize=(18,9))
plt.plot(df['Year'],df['Yearly_Expenses'],label='Yearly Expenses')
plt.plot(df['Year'],df['Yearly_Withdrawl_Amount'],label='Yearly Withdrawl Amount')
plt.title('Years Until Retirement')
plt.xlabel('Year')
plt.ylabel('INR')
plt.xticks(df['Year'],rotation=45)
plt.legend()
plt.show()
```
