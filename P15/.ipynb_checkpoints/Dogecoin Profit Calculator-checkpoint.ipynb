```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df =pd.read_csv('DOGE-USD.csv')
df
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
```

```python
df
```

```python
# ask the user the amount of money invested and date that they invested
amount_invested = input('Amonut Invested in $')
print(amount_invested)
invest_date = input('Enter date you have invested')
print(invest_date)
```

```python
# get low and high price of the day user has invested
#create variables to store the low and high price column names
col1 = 'Low'
col2 = 'High'

price1 = df[col1][invest_date]
price2 = df[col2][invest_date]

#compute quantity of asset that user would have
quantity1 = int(amount_invested) / price1
quantity2 = int(amount_invested) / price2

#compute the profit 
profit1 = (quantity1 * df[col1][-1]) - int(amount_invested)
profit2 = (quantity2 * df[col2][-1]) - int(amount_invested)

#print the range amount that the user would'vs made or loss
print('You would have made between $', round(profit1,2), 'and $', round(profit2,2),'as of', df['Date'][-1])
```

```python
#compute return on investment range
ROT1 = profit1/ int(amount_invested) * 100
ROT2 = profit2/ int(amount_invested) * 100
print('Yoour Return on Inveestment would be between', round(ROT1,2), '% and',round(ROT2,2),'% as of', df['Date'][-1])
```
