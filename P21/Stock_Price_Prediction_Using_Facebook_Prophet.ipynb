```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('TATAMOTORS.NS.csv')
```

```python
df
```

```python
df = df[['Date',"Close"]]

df = df.rename(columns={'Date':'ds','Close':'y'})
```

```python
df
```

```python
#get the last 20  rows of data and store them into a new variable
last = df[len(df)-20:]
last
```

```python
#get all of the rows in the data set except the last 20 rows
df = df[:-20]
df
```

```python
from fbprophet import Prophet
fbp = Prophet(daily_seasonality=True)
fbp.fit(df)
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#Import facebooks prophet plotting libraries
from fbprophet.plot import plot_plotly
plot_plotly(fbp,forecast)
```

```python
# show the models prediction for 03-25-2020
forecast[forecast.ds == '03-25-2021']['yhat']
```

```python
#Show the actual value of stock for 03-25-2020
last[last.ds == '03-25-2021']['y']
```
